## Scrape 12 years of Perth's weather data from wunderground.com 

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from datetime import date, timedelta
import pandas as pd
import logging

# Configure logging
logging.basicConfig(filename='weather_scraping.log', level=logging.ERROR, format='%(asctime)s:%(levelname)s:%(message)s')

chrome_options = Options()

chromedriver_path = r"C:\Users\truon\Downloads\chromedriver.exe"
service = Service(chromedriver_path)

def create_driver():
    return webdriver.Chrome(service=service, options=chrome_options)

def scrape_weather_data(start_date, end_date):
    driver = create_driver()

    date_lst = []
    current_date = start_date
    while current_date <= end_date:
        date_lst.append(current_date.strftime('%Y-%m-%d'))
        current_date += timedelta(days=1)

    all_data_dict = {}

    try:
        for date in date_lst:
            try:
                url = f"https://www.wunderground.com/history/daily/au/perth/YPPH/date/{date}"
                driver.get(url)
                
                
                try:
                    WebDriverWait(driver, 150).until(EC.presence_of_element_located((By.CSS_SELECTOR, 'table.mat-table')))
                except:
                    logging.error(f"Timeout error for Perth on date {date}, skipping this date.")
                    continue

                table = driver.find_element(By.CSS_SELECTOR, 'table.mat-table')

                data = []

                rows = table.find_elements(By.TAG_NAME, 'tr')
                for row in rows[1:]:
                    columns = row.find_elements(By.TAG_NAME, 'td')

                    if len(columns) >= 8:
                        time = columns[0].text.strip()
                        data.append({
                            "Date": date,
                            "Time": time,
                            "Temperature": columns[1].text.strip(),
                            "Dew Point": columns[2].text.strip(),
                            "Humidity": columns[3].text.strip(),
                            "Wind Speed": columns[5].text.strip(),
                            "Wind Gust": columns[6].text.strip(),
                            "Pressure": columns[7].text.strip(),
                        })

                all_data_dict[date] = data
            except Exception as e:
                logging.error(f"An error occurred for Perth on date {date}: {e}")
    except Exception as e:
        logging.error(f"An overall error occurred: {e}")
    finally:
        driver.quit()

    final_data = []

    for date, data in all_data_dict.items():
        final_data.extend(data)

    df = pd.DataFrame(final_data)

    return df
```

```python
start_date = date(2012, 1, 1)
end_date = date(2024, 8, 1)

df = scrape_weather_data(start_date, end_date)
```

## Use StandardScaler to scale the data, since some of the algorithms are distance-based

```python
from sklearn.preprocessing import StandardScaler
temp = df["Temperature"]
temp = temp.sort_index().loc[~temp.index.duplicated(keep='first')]

# Resample the data to a 30-minute interval, filling any gaps
temp = temp.resample('30T').interpolate(method='time')

# Separate features and target variable
features = df[['Dew Point', 'Humidity', 'Wind Speed', 'Wind Gust', 'Pressure']]
target = df['Temperature']

# Standardize the features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Convert back to DataFrame
features_scaled = pd.DataFrame(features_scaled, index=features.index, columns=features.columns)

# Combine standardized features with the target variable
df = pd.concat([features_scaled, target], axis=1)

df.head()
```

## Splitting Training and Testing datasets

```python
import sklearn 
from sklearn.model_selection import train_test_split
features = df.drop(columns = "Temperature", axis = 1)
target = df["Temperature"]
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)


print("Training Features Shape:", X_train.shape)
print("Testing Features Shape:", X_test.shape)
print("Training Target Shape:", y_train.shape)
print("Testing Target Shape:", y_test.shape)
```

## Applying Random Forests to predict the Temperature based on the rest of columns

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

# Initialize and train the RandomForestRegressor
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

# Evaluate the model
train_score = rf.score(X_train, y_train)
test_score = rf.score(X_test, y_test)

# Calculate the mean squared error
y_train_pred = rf.predict(X_train)
y_test_pred = rf.predict(X_test)
train_mse = mean_squared_error(y_train, y_train_pred)
test_mse = mean_squared_error(y_test, y_test_pred)

print(f"Random Forest Training Score: {train_score}")
print(f"Random Forest Testing Score: {test_score}")
print(f"Random Forest Training MSE: {train_mse}")
print(f"Random Forest Testing MSE: {test_mse}")
```

## Applying K-Nearest Neighbors
_As we see the importances of features above, I now use KNN but with different features each time, with the n-neighbors from 1 to 10 and also plot the training and testing score of each model_


```python
from sklearn.neighbors import KNeighborsRegressor

fig, axs = plt.subplots(3, 2, figsize=(12, 13))  

feature_lst = ["Wind Gust", "Wind Speed", "Pressure"]
features_drop = []

for i in range(0, len(feature_lst)):
    features_drop.append([feature_lst[i]])
    for j in range(i + 1, len(feature_lst)):
        features_drop.append([feature_lst[i], feature_lst[j]])

for idx, cols in enumerate(features_drop):
    features = df.drop(columns=["Temperature"] + cols, axis=1)
    target = df["Temperature"]
    X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
    
    n_neighbors = range(1, 11)
    train_score = []
    test_score = []
    
    for n in n_neighbors:
        knn = KNeighborsRegressor(n_neighbors=n)
        knn.fit(X_train, y_train)
        train_score.append(knn.score(X_train, y_train))
        test_score.append(knn.score(X_test, y_test))

    ax = axs[idx // 2, idx % 2]  # Correct subplot indexing

    ax.plot(n_neighbors, train_score, label="Train score", marker="o")
    ax.plot(n_neighbors, test_score, label="Test score", marker="x")
    ax.set_xlabel("N_neighbors")
    ax.set_ylabel("Accuracy")
    ax.set_title(f"Excluding {cols}")
    ax.legend()
    ax.grid()

plt.tight_layout()
plt.show()
```

_We can see that all of those versions yield significant scores, but how about we drop all of those 3 insignificant estimators which are Wind Gust, Wind Speed and Pressure_

```python
features = df[["Dew Point","Humidity"]]
target = df["Temperature"]
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

test_score = [] 
train_score = []
for n in range(1,11):

    knn = KNeighborsRegressor(n_neighbors = n)
    knn.fit(X_train, y_train)
    
    train_score.append(knn.score(X_train, y_train))
    test_score.append(knn.score(X_test, y_test))
print(train_score)
print(test_score)

n_neighbors = range(1,11)
plt.plot(n_neighbors, train_score, label = "Train score", marker = "o" )
plt.plot(n_neighbors, test_score, label = "Test score", marker = "x")
plt.xlabel("N_neighbors")
plt.ylabel("Accuracy")
plt.legend()
plt.grid()
```

## Applying the last algorithm: XGBoost

```python
from sklearn.model_selection import GridSearchCV
from xgboost import XGBRegressor

features = df.drop(columns = "Temperature", axis = 1)
target = df["Temperature"]
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

param_grid = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 4, 5, 6],
    'subsample': [0.8, 0.9, 1.0],
    'colsample_bytree': [0.8, 0.9, 1.0]
}

# Initialize the model
xgb_model = XGBRegressor(objective='reg:squarederror')

# Initialize GridSearchCV
grid_search = GridSearchCV(estimator=xgb_model, param_grid=param_grid, scoring='neg_mean_squared_error', cv=5, verbose=1)


grid_search.fit(X_train, y_train)


print(f"Best parameters: {grid_search.best_params_}")

# Use the best model
best_xgb_model = grid_search.best_estimator_
y_test_pred_xgb = best_xgb_model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_test_pred_xgb)
print(f"Mean Squared Error (MSE): {mse}")
```

### Checking for stationary by ADF test

```python
from statsmodels.tsa.stattools import adfuller
temp = df["Temperature"]

result = adfuller(temp)

output = pd.Series(result[0:4], index = ["Test Statistics", "P-value", "Number of Lags Used", "Number of Observations Used"])
for key, value in result[4].items():
    output["{} Critical Values : ".format(key)] = value
print(output)
```

_The p-value is extremely small (2.419841e-30), far below the typical significance levels of 0.01, 0.05, and 0.10_

##### Based on these results, we can conclude that there is strong evidence to reject the null hypothesis of a unit root, indicating that the time series is stationary

### ACF and PACF plots

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plot_acf(temp)
plot_pacf(temp)
```

#### Choosing parameters
+ Parameter(p): A strong spike at lag 1 in the PACF suggests that an AR(1) model might be appropriate
+ Parameter(d): The data is already stationary, then the differencing parameter 𝑑 should be 0
+ Parameter (q): The sinusoidal pattern suggests that a higher-order MA component might be necessary, so I choose 3

### Predicting temperature of every 2 hours 

```python
import pandas as pd
from statsmodels.tsa.statespace.sarimax import SARIMAX
import matplotlib.pyplot as plt

temp.index = pd.to_datetime(temp.index)

temp.sort_index(inplace=True)

# Resample data to 2-hour intervals since 1H takes too long to compute
temp = temp.resample('2H').mean()

train = temp.loc['2018-07-01':'2024-07-24']
test = temp.loc['2024-07-25':'2024-07-30']

model = SARIMAX(train, order=(1, 0, 3), seasonal_order=(0, 1, 1, 24))
model_fit = model.fit(disp=False)

forecast = model_fit.get_forecast(steps=72)
forecasted_values = forecast.predicted_mean

plt.figure(figsize=(12, 6))
plt.plot(forecasted_values.index, forecasted_values, label='Forecasted Temperature')
plt.plot(test.index, test, label='Actual Temperature', color='orange')
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.title('Temperature Forecast vs Actual')
plt.legend()
plt.show()
```

*Quite impressive as we are predicting the temperature of each day for every 2 hours interval, but let's see the mean square error*

```python
from sklearn.metrics import mean_squared_error
print(f" Root Mean Square Error: {mean_squared_error(forecasted_values, test) ** 0.5}")
```
